```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
#Read
hero_info = pd.read_csv('Data/superhero_info.csv')
hero_info.head()
```

```python
#Read
powers = pd.read_csv('Data/superhero_powers.csv')
powers.head()
```

```python
# split the 'Hero|Publisher' into 2 columns
hero_info[['Hero','Publisher']] = hero_info['Hero|Publisher'].str.split('|',expand=True)
hero_info.head(2)
```

```python
## drop the original column 
hero_info = hero_info.drop(columns=['Hero|Publisher'])
hero_info.head(2)
```

```python
## examining a single value from the coordinates col
measurement = hero_info.loc[0,"Measurements"]
print(type(measurement))
measurement
```

```python
measurement = measurement.replace("'",'"')
measurement
```

```python
## now we can use json.loads
fixed_measurement = json.loads(measurement)
print(type(fixed_measurement))
fixed_measurement
```

```python
## use .str.replace to replace all single quotes
hero_info['Measurements'] = hero_info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
hero_info['Measurements'] = hero_info['Measurements'].apply(json.loads)
hero_info['Measurements'].head()
```

```python
## check a single value after transformation
test_measurement = hero_info.loc[0, 'Measurements']
print(type(test_measurement))
test_measurement
```

```python
#create unique columnds for each dictionary key/value in 'Measurements'
height_weight = hero_info['Measurements'].apply(pd.Series)
height_weight
```

```python
#Add New Columns
hero_info = pd.concat((hero_info, height_weight), axis = 1)
hero_info.head(2)
```

```python
##Dro Original Column
hero_info = hero_info.drop(columns=['Measurements'])
hero_info.head(2)
```

```python
#Rename
powers.rename({'hero_names': 'Hero', 'Powers':'Powers'}, axis = 1, inplace = True)
powers.head()
```

```python
powers.loc[1,'Powers']
```

```python
powers['temp_split'] = powers['Powers'].str.split(',',expand = False)

powers['temp_split'].head()
```

```python
exploded = powers.explode('temp_split')

exploded
```

```python
#Unique Values
cols_to_make = exploded['temp_split'].dropna().unique()
```

```python
for col in cols_to_make:
    powers[col] = powers['Powers'].str.contains(col)
```

```python
powers.head()
```

```python
powers = powers.drop(columns = ['Powers', 'temp_split'])
powers.head()
```

```python
## Merging data frames on Hero
superheros = pd.merge(hero_info, powers, left_on = 'Hero', right_on = 'Hero')
superheros
```

```python
# split the 'Height' into 2 columns
superheros['Height'].str.split(' ',expand=True)
superheros.head()
```

```python
#Save and rename the Height
superheros[['Height','Height_UOM']] = superheros['Height'].str.split(' ',expand=True)
superheros.head()
```

```python
#Save and rename the Weight
superheros[['Weight','Weight_UOM']] = superheros['Weight'].str.split(' ',expand=True)
superheros.head()
```

```python
superheros.info(['Height', 'Weight'])
```

```python
#Change Data Type
superheros['Height']=superheros['Height'].astype(float)
```

```python
#Change Data Type
superheros['Weight']=superheros['Weight'].astype(float)
```

```python
superheros.info(['Height','Weight'])
```

```python
#Final
superheros.head()
```

```python
#Compare the average weight of super powers who have Super Speed to those who do not.
superheros.groupby('Super Speed')['Weight'].mean()
```

```python
#What is the average height of heroes for each publisher?
superheros.groupby('Publisher')['Height'].mean()
```
